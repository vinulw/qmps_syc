```python
def fit_ps(p, xs):
    ys = np.zeros(xs.shape)
    for i, p in enumerate(reversed(p)):
        ys += p*xs**i
    return ys
```

```python
def calculate_ratio_error(Ci, sigma_Ci):
    ratios = Ci[1:]/ Ci[:-1]
    return ratios * np.sqrt( (sigma_Ci[1:]/Ci[1:])**2 + (sigma_Ci[:-1]/Ci[:-1])**2 )
```
